````python
import requests
import time
time.sleep(0.2)
from bs4 import BeautifulSoup
import pandas as pd
KEYWORDS = ['python', 'парсинг']
````

````python
def pages():
    list_pages = []
    req = requests.get('https://habr.com/ru/all/')
    soup = BeautifulSoup(req.text, 'html.parser')
    time.sleep(0.2)
    
    pages = int(soup.find('a',class_= \
    'toggle-menu__item-link toggle-menu__item-link_pagination toggle-menu__item-link_bordered').get('href')[-3:-1])         
    for page in range(1, pages +1):
        list_pages += 'https://habr.com/ru/all/page' + str(page) + '/',
    return list_pages 
pages()
````

````python
def df_pages():
    df = pd.DataFrame()
    KEYWORDS = ['python', 'парсинг', 'задача']
    for url_page in pages():
        req = requests.get('https://habr.com/ru/all/')
        soup = BeautifulSoup(req.text, 'html.parser')
        posts = soup.find_all('article', class_='post_preview')
        
        for post in posts:
            text = post.find('a', class_='post__title_link')
            link = text.get('href')
            title = text.get_text()
            date = post.find('span', class_='post__time').text
#             print(text.text, text.attrs.get('href'))
            for key in KEYWORDS: 
                if key in title:
                    row = {'date': date, 'title': title, 'post':link }
                    df = pd.concat([df, pd.DataFrame([row])]) 
        return(df)
df_pages()
````

В итоге должен формироваться датафрейм со столбцами: <почта> - <дата утечки> - <источник утечки> - <описание утечки>.

````python
import json
````

````python
EMAIL = ['xxx@x.ru', 'yyy@y.com']
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
head = {'Vaar-Header-App-Product': 'hackcheck-web-avast','Vaar-Version': '0'}
adresses = {'emailAddresses': EMAIL}
````

````python
res = requests.post(url, data = json.dumps(adresses), headers = head)
res.json()
````

````python
df = pd.DataFrame(res.json()['breaches'])\
        .transpose()\
        .reset_index() 
df_final = df[['site', 'description', 'publishDate']]
df_final
````
